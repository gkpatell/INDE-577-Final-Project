```python
mypic_red = mypic[:,:,0] #picks out the red/first number in each RGD triple
mypic_red
```

```python
mypic_blue = mypic [:,:,1]
mypic_green = mypic [:,:,2]
```

```python
U_red, s_red, V_red = np.linalg.svd(mypic_red) #Computing SVD of the matrix of the 1080 X 1920 matrix of red component
U_red, s_red, V_red
```

```python
U_blue, s_blue, V_blue, = np.linalg.svd(mypic_blue)
U_green, s_green, V_green = np.linalg.svd(mypic_green)
```

```python
rank = 5 #The rank is the number of singular values we use to compress our image
compress_red = U_red[:, :rank] @ np.diag(s_red[:rank])@V_red[:rank, :] # Use @ instead of * for matrix built in Python!
compress_red
```

```python
compress_red = compress_red.astype(int)
compress_red
```

```python
def compress(rank):
    compress_red = U_red[:, :rank] @ np.diag(s_red[:rank])@V_red[:rank, :]
    compress_red = compress_red.astype(int)

    compress_blue = U_blue[:, :rank] @ np.diag(s_blue[:rank])@V_blue[:rank, :]
    compress_blue = compress_blue.astype(int)

    compress_green = U_green[:, :rank] @ np.diag(s_green[:rank])@V_green[:rank, :]
    compress_green = compress_green.astype(int)

    compressed_array = np.stack((compress_red, compress_blue, compress_green), axis = 2)

    rows,cols = np.shape(mypic_red)
    compression_ratio = 100*rank*(rows + cols + 1) / (rows * cols)
    print("Compression Percent", compression_ratio)
    plt.imshow(compressed_array)
```

```python
compress(5)
```

```python
compress(20)
```

```python
compress(100)
```
